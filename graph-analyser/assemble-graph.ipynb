```python
from typing import List


game_node_ids: set[int] = set()
user_node_ids: set[int] = set()
edges_details_by_review_id: dict[int, dict] = {}
edges_review_id_user_id_pairs_by_game_id: dict[int, set[tuple[int, int]]] = {}
edges_review_id_game_id_pairs_by_user_id: dict[int, set[tuple[int, int]]] = {}
```

```python
import json
with open('../steam-applist-scraper/filtered_apps_ids.json', 'r') as f:
  filtered_app_ids = json.load(f)
  filtered_app_ids = [int(appid) for appid in filtered_app_ids]

game_nodes_ids = set(filtered_app_ids)
```

```python
import os

directory = '../steam-review-scraper/data'
game_ids_with_review_file = sorted([int(filename.split("_")[1].split(".")[0]) for filename in os.listdir(directory)], reverse=True)

print(game_ids_with_review_file)
```

```python
import gc
from tqdm import tqdm

for appid in tqdm(game_ids_with_review_file):
  gc.collect()

  with open(f'../steam-review-scraper/data/review_{appid}.json', 'r') as f:
    reviews = json.load(f)
    for review_id, review in reviews['reviews'].items():
      user_id = review['author']['steamid']
      
      user_node_ids.add(int(user_id))
      game_node_ids.add(int(appid))
      edges_details_by_review_id[int(review_id)] = {k: review[k] for k in ('voted_up', 'received_for_free')}
      
      edges_game_ids_by_user_id.setdefault(int(user_id), set()).add(int(appid))
      edges_review_id_user_id_pairs_by_game_id.setdefault(int(appid), set()).add((int(review_id), int(user_id)))
```

```python
with open('game_node_ids.json', 'w') as f:
  json.dump(list(game_node_ids), f)

with open('user_node_ids.json', 'w') as f:
  json.dump(list(user_node_ids), f)

with open('edges_user_ids_by_game_id.json', 'w') as f:
  json.dump({str(k): list(v) for k, v in edges_user_ids_by_game_id.items()}, f)

with open('edges_game_ids_by_user_id.json', 'w') as f:
  json.dump({str(k): list(v) for k, v in edges_game_ids_by_user_id.items()}, f)
```
